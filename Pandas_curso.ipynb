### Lectura csv

Un archivo CSV (Comma Separated Values o Valores Separados por Comas) es un archivo de texto plano que almacena datos estructurados en forma de tabla, donde cada línea representa una fila y las comas (o puntos y coma) separan las columnas. Es un formato universal, ligero y compatible con hojas de cálculo como Excel y bases de datos. 

La lectura de csv contiene los siguientes campos

```python
df = pd.read_csv(
    "Ecommerce.csv",
    sep = ",",           # separador
    encoding = "utf-8",  # codificación
    header = 0,          # fila de encabezados 0 si la tiene, 1 si NO la tiene
    index_col = None     # Se coloca la columna indice
)

print(df)
```

### Outliers

Un outlier, o dato atípico, es una observación en un conjunto de datos que se desvía significativamente del comportamiento general del resto. Son valores extremos, "ovejas negras" que rompen el patrón, indicando anomalías, errores de medición o eventos inusuales. Su identificación es crucial para evitar sesgos en promedios y análisis.

```python
# Crear un dataframe
# Crear dataset con 50 registros y outliers intencionales
np.random.seed(42)

# Datos normales
edades_normales = np.random.normal(35, 10, 45)  # 45 datos normales
edades_outliers = np.array([15, 16, 65, 70, 120])  # 5 outliers

salarios_normales = np.random.normal(50000, 15000, 45)
salarios_outliers = np.array([5000, 150000, 200000, 250000, 500000])

df = pd.DataFrame({
    'id': range(1, 51),
    'edad': np.concatenate([edades_normales, edades_outliers]),
    'salario': np.concatenate([salarios_normales, salarios_outliers]),
    'puntuacion': np.random.randint(1, 100, 50),
    'habitaciones': np.random.randint(1, 5, 50)
})

# Añadir un outlier en habitaciones
df.loc[10, 'habitaciones'] = 15
```

```python
# Rango intercuartilico metodo

def detectar_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    
    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]
    
    print(f"\n   Columna: {columna}")
    print(f"   Q1 (25%): {Q1:.2f}")
    print(f"   Q3 (75%): {Q3:.2f}")
    print(f"   IQR: {IQR:.2f}")
    print(f"   Límite inferior: {limite_inferior:.2f}")
    print(f"   Límite superior: {limite_superior:.2f}")
    print(f"   Outliers detectados: {len(outliers)}")
    print(f"   Valores outliers: {outliers[columna].tolist()}")
    
    return outliers


    
```

```python
# Ejemplo de aplicacion
# Aplicar a todas las columnas numéricas
columnas_numericas = ['edad', 'salario', 'puntuacion', 'habitaciones']
outliers_iqr = {}

for col in columnas_numericas:
    outliers_iqr[col] = detectar_outliers_iqr(df, col)
```

```python
# Metodo z-score

def detectar_outliers_zscore(df, columna, umbral=3):
    media = df[columna].mean()
    std = df[columna].std()
    z_scores = np.abs((df[columna] - media) / std)
    outliers = df[z_scores > umbral]
    
    print(f"\n  Columna: {columna}")
    print(f"   Media: {media:.2f}")
    print(f"   Desviación estándar: {std:.2f}")
    print(f"   Umbral Z-score: {umbral}")
    print(f"   Outliers detectados: {len(outliers)}")
    print(f"   Valores outliers: {outliers[columna].tolist()}")
    
    # Mostrar Z-scores de los outliers
    if len(outliers) > 0:
        print(f"   Z-scores: {z_scores[outliers.index].tolist()}")
    
    return outliers
```

```python
# Ejemplo de aplicacion
# Aplicar a todas las columnas numéricas
columnas_numericas = ['edad', 'salario', 'puntuacion', 'habitaciones']
outliers_iqr = {}

for col in columnas_numericas:
    outliers_iqr[col] = detectar_outliers_zscore(df, col,umbral = 3)
```

## <font color = darkblue> Normalización, escalamiento y estandarización

En Machine Learning y ciencia de datos, los datos deben estar en un rango o formato uniforme para garantizar que los algoritmos funcionen correctamente. Aquí es donde entran en juego las técnicas de normalización, escalamiento y estandarización. Vamos a explicar cada una:

<center> <img src = "Normalizacion_estandarizacion_escalamiento.png" width="600"></center>

### Normalización
La normalización consiste en reescalar los valores de las características (features) para que queden en un rango específico, generalmente entre 0 y 1. Es útil cuando los datos no siguen una distribución normal o cuando se quiere limitar los valores.


<center> $$x' = \frac{x - x_{min}}{x_{max} - x_{min}}$$</center>

donde $x$ es el valor original, $x_{min}$ es el valor mínimo de la característica, $x_{max}$ es el valor máximo de la característica, y $x'$ es el valor normalizado.

Esta técnica se utiliza cuando los datos tienen escalas muy diferentes y no tienen una distribución normal. También cuando se requiere que los valores estén en un rango definido.


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Datos originales
datos = [[1], [2], [3], [4], [5]]

# Normalización
scaler = MinMaxScaler()
datos_normalizados = scaler.fit_transform(datos)
datos_normalizados
```

### Escalamiento

El escalamiento ajusta los datos para que estén dentro de un rango específico, pero no necesariamente entre 0 y 1. Esto incluye técnicas como dividir entre un valor constante o trabajar con logaritmos.

<center> $$x' = \frac{x}{\max{(|x|)}}$$</center>

Se debe usar cuando la escala de los datos es importante para la interpretación, pero no necesitas que esté en un rango fijo. Y cuando algunos modelos son sensibles a las magnitudes de las características.


```python
from sklearn.preprocessing import MaxAbsScaler
```

```python
# Datos originales
datos = [[-1], [0], [1], [2], [3]]

# Escalamiento basado en el máximo absoluto
scaler = MaxAbsScaler()
datos_escalados = scaler.fit_transform(datos)
datos_escalados
```

### Estandarización

La estandarización transforma los datos para que tengan una media ($\mu$) de $0$ y una desviación estándar ($\sigma$) de $1$. Esto asegura que los datos sigan una distribución normal estándar (o lo más cercana posible).

<center> $$z = \frac{x-\mu}{\sigma}$$</center>

donde $\mu$ es la media de la característica, $\sigma$ es la desviación estándar de la característica, y $z$ es el valor estandarizado.

Se debe usar cuando los datos siguen una distribución normal o casi normal. También, cuando se utilizan modelos que asumen distribuciones normales, como la regresión logística, SVM o redes neuronales.



```python
from sklearn.preprocessing import StandardScaler
```

```python
# Datos originales
datos = [[1], [2], [3], [4], [5]]

# Estandarización
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(datos)

datos_estandarizados
```
